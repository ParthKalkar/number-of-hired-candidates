```python
#One hot encoding for rootype column
one_hot = pd.get_dummies(Trained['Sent To Training'])
Trained = Trained.drop('Sent To Training',axis = 1)
# Join the encoded df
Trained = Trained.join(one_hot)
Trained
```

```python
# df2=Trained.groupby('Position','Geo. If itÂ´s a line manager - choose School').size()
# df2 = Trained.groupby(['Position','Geo. If itÂ´s a line manager - choose School'], as_index=False)
df2=Trained.pivot_table('Yes', ['Position', 'Geo'], aggfunc='sum').reset_index()
```

```python
df2
```

```python
df_Awake=df2[df2['Position']== 'Awake']
df_Awake
```

```python
 df_Awake.plot.bar(x='Geo',y='Yes')
```

```python
#Barplot of Room Type
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,5),dpi=100),
plt.style.use('default')
ax = sns.barplot(x="Geo", y="Yes", data=df_Awake)
plt.title("Postion Awake", weight="bold", c="blue", fontsize=15)
plt.xlabel("",weight="bold",c="k")
plt.xticks(rotation=0, weight="bold")
plt.ylabel("Number of Yes)",weight= "bold",c="k")
plt.yticks(np.arange(0,2,step=1), weight="bold")
for p in ax.patches:
        ax.annotate('{:.0f}'.format(p.get_height()), (p.get_x()+0.4, p.get_height()),
                    ha='center', va='bottom',weight="bold",
                    color= 'black')
```
